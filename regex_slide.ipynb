#Regex là gì?

- regex là từ viết tắt của regular expression, đối với tiếng Việt thì còn được biểu thức chính quy
- regex là chuỗi string được <b>kết hợp</b> bởi các <b>kí tự thông thường</b> và các <b>kí tự đặc biệt</b> mà chuỗi string đó mô tả <b>"mẫu/pattern"</b> để <b>tìm kiếm kí tự hay vị trí </b> của kí tự cần tìm trong đoạn văn/ câu văn

ex: <br />
	
```Clary has 2 friends who sepnds a lot time with. Susan has 3 brother while John has 4 sisters.``` <br />
``` Today is 23rd May 2019. Tomorrow is 24th May 19.```

```Just check out links: www.amazingpics.com, www.flickr.com, www.instagram.com. They have amazing photos!```

#Ứng dụng của regex


1.   tìm kiếm và thay thế các kí tự theo 1 pattern/mẫu cụ thể nào đó
2.    xác thực chuỗi/ validate strings



#Regex characters


##Character classes

```\w\d\s```	word, digit, whitespace <br />
```\W\D\S```	not word, digit, whitespace <br />

##Special characters <br />
```^abc$```	start/end of the string<br />
```\.\*\\```	escaped special characters<br /> 
```\t\n\r```	tab, linefeed, carriage return<br />


##OR character<br />

```ab|cd```	match ab or cd <br />
```[abc]```	any of a, b, or c <br />
```[^abc]```	not a, b, or c <br />
```[a-g]```	character between a & g <br />


##Quantifiers
```a* a+ a?```	0 or more, 1 or more, 0 or 1<br />
```a{5}a{2,}```	exactly five, two or more<br />
```a{1,3}```	between one & three<br />
```a+?a{2,}?```	match as few as possible<br />



##Groups & Lookaround<br />
```(abc)```	capture group<br />
```\1```	backreference to group #1<br />
```(?:abc)```	non-capturing group<br />
```(?=abc)```	positive lookahead<br />
```(?!abc)```	negative lookahead<br />

note: metacharacters (need to be escaped) inlcudes: ```.[]{}()\^$*+?```

## Grouping and capturing

---
###Group characters: ```( )```

```	Clary has 2 friends who sepnds a lot time with. Susan has 3 brother while John has 4 sisters.```

Cần lọc ra các tên riêng: Clary, Susan, John

````python
sentence = "Clary has 2 friends who sepnds a lot time with. Susan has 3 brother while John has 4 sisters."
````

````python
re.findall('[A-Za-z]+\s\w+\s\d+\s\w+', sentence)
````

````python
re.findall('([A-Za-z]+)\s\w+\s\d+\s\w+', sentence)
````

#### Multiple groups

````python
re.findall('([A-Za-z]+)\s\w+\s(\d+)\s(\w+)', sentence)
````

````python
for match in re.finditer('([A-Za-z]+)\s\w+\s(\d+)\s(\w+)', sentence):
  print(match)
````

----
#### capture a repeated group ```(\d+)``` vs repeat a capturing group ```(\d)+```

````python
re.findall(r"(\d)+", "My lucky numbers are 8755 and 33")
````

````python
re.findall(r"(\d+)", "My lucky numbers are 8755 and 33")
````

##Alternation and non-capturing groups

###Alternation
use groups to choose betwwen optional patterns
* use ```|```


````python
my_string = "I want to have a pet. But I don't know if I want 2 cats, 1 dog or a bird."
re.findall(r"\d+\s(cat|dog|bird)", my_string)
````

````python
matches = re.finditer(r"\d+\s(cat|dog|bird)", my_string)
for match in matches:
  print(match)
````

### Non-capturing groups
match but not capture a group
* add ```?:``` : ```(?:regex)```

````python
re.findall(r"(\d+)(th|rd)", "Today is 23rd May 2019. Tomorrow is 24th May 19.")
````

````python
re.findall(r"(\d+)(?:th|rd)", "Today is 23rd May 2019. Tomorrow is 24th May 19.")
````

## Indexing Group

---
### numbered groups
17.06.2021_00.26.56_REC.png


````python
text = "Python 3.0 was released on 12-03-2008."
information = re.search('(\d{1,2})-(\d{2})-(\d{4})', text)
print("group(0): ", information.group(0))
print("group(1): ", information.group(1))
print("group(2): ", information.group(2))
print("group(3): ", information.group(3))
#print("group(4): ", information.group(4))
````

---
### named group

give a name to group

* ```(?P<name>regex)```

````python
text = "Ha Noi, 10000"
cities = re.search(r"(?P<city>[A-Za-z\s]+).*?(?P<zipcode>\d{5})", text)
cities.group("city")
````

````python
cities.group("zipcode")
````

## Lookaround
allow us to confirm that sub-pattern is ahead or behind main pattern

17.06.2021_00.46.10_REC.png

* positive and negative lookaround are non-capturing group

* positive lookaround is check that match preceded/followed text

* negative lookaround is check that match preceded/followed text


### look-ahead
17.06.2021_00.48.13_REC.png

#### positive look-ahead

````python
my_text = "tweets.txt transferred, mypass.txt transferred, keywords.txt error"
re.findall(r"\w+\.txt(?=\stransferred)", my_text)
````

negative lock-ahead

````python
my_text = "tweets.txt transferred, mypass.txt transferred, keywords.txt error"
re.findall(r"\w+\.txt(?!\stransferred)", my_text)
````

###Look-behind

17.06.2021_00.56.10_REC.png

positive look-behind

````python
my_text = "Member: Angus Young, Member: Chris Slade, Past: Malcolm Young, Past: Cliff Williams."
re.findall(r"(?<=Member:\s)\w+\s\w+", my_text)
````

negative look-behind

````python
my_text = "My white cat sat at the table. However, my brown dog was lying on the couch."
re.findall(r"(?<!brown\s)(cat|dog)", my_text)
````

#Performance

````python
import re
import time

def timed(func):
    def wrapper(*args):
        t = time.time()
        result = func(*args)
        t = time.time() - t
        print('%s took %.3f seconds.' % (func.__name__, t))
        return result
    return wrapper

regularExpression = r'\w+\s+([0-9_]+)\s+\w*'
testString = "average    2 never"

@timed
def noncompiled():
    for x in range(0,1000000):
        m = re.match(regularExpression, testString)

@timed
def compiled():
    rgx = re.compile(regularExpression)
    for x in range(0,1000000):
        m = rgx.match(testString)

@timed
def compiledInLoop():
    for x in range(0,1000000):
        rgx = re.compile(regularExpression)
        m = rgx.match(testString)

noncompiled()
compiled()
compiledInLoop()
````
