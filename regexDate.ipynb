```
sep = "[\/\._-]"
ddmm_day31_re = f"(?P<day31>(31(?={sep}(0?[13578]|1[02]))(?!{sep}11)))"
ddmm_day30_re = f"(?P<day30>(30(?!{sep}(0?2))))"

mmdd_day31_re = f"(?P<day31>(((?<=(0[13578]|1[02]){sep})|(?<=([13578]){sep}))(?<!11{sep}))31)"
mmdd_day30_re  = f"(?P<day30>(((?<!02{sep})(?<!2{sep})|(?<=12{sep}))30))"

day = "(?P<day>((?<!\d)0?[1-9]|[12][0-9]))"
month_re = "(0?[1-9]|1[012])"
year_re = "(\d{4})"

def ddmmyyyy_pattern():
  pattern = re.compile("^(%s|%s|%s)%s%s%s%s$" % (ddmm_day31_re, ddmm_day30_re, day, sep, month_re, sep, year_re))
  # print("\n ddmmyyyy_pattern: \n", pattern.pattern)
  return pattern

def mmddyyyy_pattern():
  pattern = re.compile("%s%s(%s|%s|%s)%s%s" % (month_re, sep, mmdd_day31_re, mmdd_day30_re, day, sep, year_re))
  # print("\nmmddyyyy pattern: \n", pattern.pattern)
  return pattern

def yyyymmdd_pattern():
  pattern = re.compile("^%s%s%s%s(%s|%s|%s)$" % (year_re, sep, month_re, sep, mmdd_day31_re, mmdd_day30_re, day))
  # print("\nyyyymmdd pattern: \n", pattern.pattern)
  return pattern

def yyyyddmm_pattern():
  pattern = re.compile("^%s%s(%s|%s|%s)%s%s$" % (year_re, sep, ddmm_day31_re, ddmm_day30_re, day, sep, month_re))
  # print("\nyyyyddmm pattern: \n", pattern.pattern)
  return pattern

generate_patterns = [ ddmmyyyy_pattern,  mmddyyyy_pattern,  yyyymmdd_pattern, yyyyddmm_pattern]

def test(pattern, data_test):
  for key, data in data_test.items():
    print(f"------- {key} --------")
    for date in data:
      if pattern.match(date) is None:
        print(date)

  print("-------- DONE -------")

# ddmmyyyy_data, mmddyyyy_data, yyyymmdd_data, yyyyddmm_data = create_test()
data_test = create_test()

for gen_pattern, data in list(zip(generate_patterns, data_test)):
  test(gen_pattern(), data)
```
